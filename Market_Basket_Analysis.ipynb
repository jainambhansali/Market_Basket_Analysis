```python
#@Author : Jainam
# Importing packages.
import requests
import csv
```

```python
# Download and write Training File

url = "http://kevincrook.com/utd/market_basket_training.txt"

req = requests.get(url)

openTrainingFile = open("market_basket_training.txt", "wb")

openTrainingFile.write(req.content)

openTrainingFile.close()
```

```python
# Creating Training List


trainingList = []

with open("market_basket_training.txt", encoding='UTF-8') as file:
    reader = csv.reader(file, delimiter=',')
    for row in reader:
        trainingList.append(row[1:])

file.close()

```

```python
#     Download the Training set "market_basket_test.txt"


url1 = "http://kevincrook.com/utd/market_basket_test.txt"

req1 = requests.get(url1)

openFile = open("market_basket_test.txt","wb")  

openFile.write(req1.content) 

openFile.close() 
```

```python
# Creating Testing List

testingList = []

with open('market_basket_test.txt', encoding='UTF-8') as file2:
    
    reader = csv.reader(file2, delimiter=',')
    
    for row in reader:
        
        testingList.append(row[1:])

file2.close()
```

```python
# Generate recommendation list

printTestList = []

compareList = [ 'P01', 'P02', 'P03', 'P04', 'P05', 'P06', 'P07', 'P08', 'P09', 'P10' ]

def searchTransactionList(searchList, trainingList):   
    
    transactionCount = 0
    
    for transactionList in trainingList:
        
        if( set(searchList) == set(transactionList) ):
            
            transactionCount += 1
    
    return transactionCount

```

```python
for productList in testingList:
    
    
    recommendList = []
    
    recommendList1 = []
    
    recommendList2 = []
    
    length = len(productList)
    
    if(length == 1):
        
        for sublist in compareList:  # subList here acts as Recommendation item
            
            extendedProductList = []
            
            if( productList[0] != sublist):
                
                extendedProductList.extend( [productList[0], sublist] ) 
                
                frequencyCount = searchTransactionList(extendedProductList, trainingList) 
                
                recommendList.append( [productList[0], sublist, frequencyCount] ) 
                
        recommendList.sort( key = lambda x : x[2], reverse = True )
                                           
        recomm = recommendList[0][1]
                                           
        printTestList.append( [productList[0], recomm] )    
    
    # For Test Items with 2 Products in list
    elif(length == 2):
        
        if( 'P04' in productList ):
            
            productList.remove('P04')
        
        elif( 'P08' in productList ):
            
            productList.remove('P08')
            
        if( len(productList) == 2 ):    
        
            for sublist1 in compareList:  # subList here acts as Recommendation item
            
                extendedProductList1 = []
                
                if( productList[0] != sublist1 and productList[1] != sublist1):
                    
                    extendedProductList1.extend( [ productList[0], productList[1], sublist1 ] )  
                                               
                    frequencyCount1 = searchTransactionList(extendedProductList1, trainingList)
        
                    recommendList1.append([productList[0], productList[1], sublist1, frequencyCount1])                            
            
            recommendList1.sort( key = lambda x : x[3], reverse = True )
            
            recomm1 = recommendList1[0][2]
            
            printTestList.append( [productList[0], recomm1] )
            
        # After Removing 'P04' OR 'P08' from 2 Items baskets.
        elif( len(productList) == 1 ):
            
            for sublist1 in compareList:  # subList here acts as Recommendation item
            
                extendedProductList1 = []
                
                if( productList[0] != sublist1 ):
                    
                    extendedProductList1.extend( [ productList[0], sublist1 ] )  
                                               
                    frequencyCount1 = searchTransactionList(extendedProductList1, trainingList)
        
                    recommendList1.append([productList[0], sublist1, frequencyCount1])                            
            
            recommendList1.sort( key = lambda x : x[2], reverse = True )
            
            recomm1 = recommendList1[0][1]
            
            printTestList.append( [productList[0], recomm1] )
            
    # For Items Baskets Product List = 3
    elif(length == 3):
        
        #Removing Products 'P04' and 'P08'
        
        if( 'P04' in productList and 'P08' in productList):
            
            productList.remove('P04')
            
            productList.remove('P08')
        
        elif( 'P08' in productList ):
            
            productList.remove('P08')
        
        elif( 'P04' in productList ):
            
            productList.remove('P04')
        
        if( len(productList) == 3 ):
        
            for sublist2 in compareList:  # subList here acts as Recommendation item
            
                extendedProductList2 = []
                
                if( productList[0] != sublist2 and productList[1] != sublist2 and productList[2] != sublist2 ):
                    
                    extendedProductList2.extend( [ productList[0], productList[1], productList[2], sublist2 ] ) 
                    
                    frequencyCount2 = searchTransactionList(extendedProductList2, trainingList)
                    
                    recommendList2.append( [productList[0], productList[1], productList[2], sublist2, frequencyCount2] ) 
                    
            recommendList2.sort( key = lambda x : x[4], reverse = True )
            
            recomm2 = recommendList2[0][3]
            
            printTestList.append( [productList[0], recomm2] )        
            
       # After Removing 'P04' OR 'P08' from 2 Items baskets.
        elif( len(productList) == 2 ):  
            
            for sublist2 in compareList:  # subList here acts as Recommendation item
            
                extendedProductList2 = []
            
                if( productList[0] != sublist2 and productList[1] != sublist2):
                        
                        extendedProductList2.extend( [ productList[0], productList[1], sublist2 ] ) 
                                                   
                        frequencyCount2 = searchTransactionList(extendedProductList2, trainingList)
            
                        recommendList2.append([productList[0], productList[1], sublist2, frequencyCount2])                            
                
            recommendList2.sort( key = lambda x : x[3], reverse = True )
                
            recomm2 = recommendList2[0][2]
                
            printTestList.append( [productList[0], recomm2] )
        
        # After Removing 'P04' OR 'P08' from 2 Items baskets.    
        elif( len(productList) == 1 ):     
            
            for sublist2 in compareList:  # subList here acts as Recommendation item
                
                extendedProductList2 = []
                
                if( productList[0] != sublist2  ):
                    
                    extendedProductList2.extend( [ productList[0], sublist2 ] ) 
                    
                    frequencyCount2 = searchTransactionList(extendedProductList2, trainingList)
                    
                    recommendList2.append( [productList[0], sublist2, frequencyCount2] ) 
                    
            recommendList2.sort( key = lambda x : x[2], reverse = True )
            
            recomm2 = recommendList2[0][1]
            
            printTestList.append( [productList[0], recomm2] )  
        
    else: 
        continue
```

```python
#Creating Recommendation File

transactionNumber = [ "{0:03}".format(i) for i in range(1,101) ]

with open('market_basket_recommendations.txt', mode='wt') as outputRecommFile: 
    
    for i in range(0, len(printTestList)):        
        
        if ( i <= len(printTestList) ):
                       
            writeLine = str(transactionNumber[i]) + ',' + str(printTestList[i][1]) +'\n'            
        
        else:            
            writeLine = str(transactionNumber[i]) + ',' + str(printTestList[i][1])     
            
        outputRecommFile.write(writeLine)

outputRecommFile.close()

```
